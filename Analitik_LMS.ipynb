```python
from google.colab import files

uploaded = files.upload()

import pandas as pd

# Load dataset
file_path = "data_quiz.xlsx"  # Sesuaikan dengan nama file yang diunggah
xls = pd.ExcelFile(file_path)

# Baca sheet "Gabungan"
df = pd.read_excel(xls, sheet_name="Gabungan")

# Tampilkan beberapa baris pertama
df.head()
```

```python
# Mengecek jumlah missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

# Menghapus baris dengan missing values
df_cleaned = df.dropna()

# Cek kembali jumlah data setelah pembersihan
print("Jumlah data setelah pembersihan:", df_cleaned.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot nilai Quiz berdasarkan kelas
plt.figure(figsize=(6,4))
sns.boxplot(x="Kelas", y="Quiz", data=df_cleaned)
plt.title("Boxplot Nilai Quiz Berdasarkan Kelas")
plt.show()
```

```python
# Boxplot nilai Quiz berdasarkan gender
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Quiz", data=df_cleaned)
plt.title("Boxplot Nilai Quiz Berdasarkan Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Quiz", y="IPS2", data=df_cleaned)
plt.title("Scatter Plot Nilai Quiz vs IPS Semester 2")
plt.xlabel("Nilai Quiz")
plt.ylabel("Nilai IPS Semester2")
plt.show()
```

```python
# Menghitung korelasi antara nilai Quiz dan IPS2
correlation = df_cleaned[["Quiz", "IPS2"]].corr()

# Menampilkan hasil korelasi
print(correlation)
```

```python
# Histogram nilai Quiz
plt.figure(figsize=(6,4))
sns.histplot(df_cleaned["Quiz"], bins=10, kde=True)
plt.title("Histogram Nilai Quiz")
plt.show()
```

```python
# Histogram nilai IPS
plt.figure(figsize=(6,4))
sns.histplot(df_cleaned["IPS2"], bins=10, kde=True)
plt.title("Histogram Nilai IPS Semester2")
plt.show()
```

```python
mean_quiz = df_cleaned["Quiz"].mean()
std_quiz = df_cleaned["Quiz"].std()

mean_ips = df_cleaned["IPS2"].mean()
std_ips = df_cleaned["IPS2"].std()

print(f"Rata-rata Nilai Quiz: {mean_quiz:.2f}, Standar Deviasi: {std_quiz:.2f}")
print(f"Rata-rata Nilai IPS Semester 2: {mean_ips:.2f}, Standar Deviasi:{std_ips:.2f}")
```

```python
# Boxplot nilai IPS2 berdasarkan kelas
plt.figure(figsize=(6,4))
sns.boxplot(x="Kelas", y="IPS2", data=df_cleaned)
plt.title("Boxplot Nilai IPS Semester 2 Berdasarkan Kelas")
plt.show()
```

```python
# Boxplot nilai IPS2 berdasarkan gender
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="IPS2", data=df_cleaned)
plt.title("Boxplot Nilai IPS Semester 2 Berdasarkan Gender")
plt.show()
```
